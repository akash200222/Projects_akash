```python
plt.pie(data["match"].value_counts(),labels=data["match"].value_counts().index, autopct="%0.1f%%",explode=[0.1,0,0])
plt.show()
# This chart shows that Virat Kohli has played most ODI matches.
```

```python
# create count plot for analysis the how many grounds did Virat Kohli play matches
plt.figure(figsize=(18,5))
sns.countplot(data["ground"])
plt.xticks(rotation=80)
plt.show()
#This chat shows that Virat Kohli has played maximum number of matches in Mirpur.
```

```python
# in this countplot shows virat kohli has played against which opeanent
plt.figure(figsize=(10,5))
sns.countplot(data["opponent"])
plt.xticks(rotation=80)
plt.show()
#This chart shows that Virat Kohli has played the maximum number of matches against England.
```

```python
# create barplot for analysis the In which ground did Virat Kohli score how many runs.
plt.figure(figsize=(18,15))
sns.barplot(x="ground",y="runs", data=data)
plt.xticks(rotation=100)
plt.show()
```

```python
#this barplot shows Virat Kohli has scored how many runs against which opponent
plt.figure(figsize=(7,5))
sns.barplot(x="opponent",y="runs" ,  data=data)
plt.xticks(rotation=100)
plt.show()
#This chart shows that Virat Kohli scored the most runs against Afghanistan.
```

```python
#This barplot divides the number of matches in which Virat Kohli has scored the most runs.
plt.figure(figsize=(7,5))
sns.barplot(x="match",y="runs" ,data=data)
plt.xticks(rotation=100)
plt.show()
#This bar chart shows that Virat Kohli has scored the most runs in ODIs and the least in T20s.
```

```python
# most runs is matches 
data1 = data.groupby(['runs'],as_index =False)['match'].sum().sort_values(by = 'runs',ascending = False).head(5)
plt.figure(figsize=(10,5))
sns.countplot(data=data1, x = "match" ,hue = "runs")
plt.show()
#This column chart shows the top 5 batting performances of Virat Kohli
```
